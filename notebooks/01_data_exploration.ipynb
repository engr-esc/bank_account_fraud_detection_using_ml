##**Abstract**

This capstone develops an explainable machine-learning workflow for prioritizing potentially fraudulent bank-account applications. The Bank Account Fraud Dataset Suite - NeurIPS 2022 was used with Base.csv as the primary development dataset and Variant II.csv as an external robustness stress test. Both datasets contain 1,000,000 records, 32 original variables, and a fraud prevalence of 1.103%, making the problem a highly imbalanced binary-classification task.


The analysis followed a temporal design: months 0-5 were used for training, month 6 for validation and threshold selection, and month 7 for final testing. Logistic Regression, Decision Tree, Random Forest, and XGBoost were tuned using PR-AUC/Average Precision. XGBoost achieved the strongest cross-validated PR-AUC (0.140452), but Logistic Regression generalized better to the independent validation month and was selected as the provisional champion with validation PR-AUC 0.164799. On the Base month 7 test set, it achieved PR-AUC 0.198539, ROC-AUC 0.890668, and captured 67.02% of fraud within the highest-risk 10% of applications. The same model achieved PR-AUC 0.201282 and top-10% recall 66.46% on Variant II without retraining.


Explainability and responsible-AI analysis identified material differences across age, income, employment, and housing groups, as well as probability-calibration and threshold-transfer concerns. The findings support use of the model as a human-supervised review-prioritization tool, not as an automated rejection mechanism. The project concludes with mitigation experiments, governance recommendations, reproducible model artifacts, and a prioritized improvement roadmap.


##**Executive Summary**

The project demonstrates that a relatively transparent Logistic Regression model can provide strong fraud-ranking performance while remaining easier to explain and govern than more complex alternatives. Under the selected 10% review-capacity policy, the final model concentrated roughly two-thirds of detected fraud into one-tenth of applications and produced an illustrative net benefit of 6.42 million under the stated cost assumptions.


The results also show that strong ranking performance does not automatically imply calibrated probabilities or equal outcomes across groups. Variant II preserved PR-AUC and top-10% recall, but the unchanged threshold generated a higher false-positive rate and weaker calibration. Age and socioeconomic proxy groups displayed material error-rate differences, and the tested reweighting strategy reduced some selection-rate gaps while worsening equal-opportunity and predictive performance. These findings justify a human-in-the-loop operating model, periodic recalibration, subgroup monitoring, and external validation before deployment.
